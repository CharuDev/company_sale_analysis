```python
import pandas as pd
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
data= pd.read_csv("file.csv", encoding= 'unicode_escape')
```

```python
data
```

```python
data.shape
```

```python
data.info()
```

```python
data.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
pd.isnull(data).sum()
```

```python
data.dropna(inplace=True)
```

```python
data['Amount']=data['Amount'].astype('int')
```

```python
data['Amount'].dtypes
```

```python
#rename column
data.rename(columns={'Amount':'paisa'})
```

```python
data.describe()
```

```python
# plotting a bar chart for Gender and it's count
value=sns.countplot(x='Gender',data=data)
for bars in value.containers:
    value.bar_label(bars)
```

```python
# plotting a bar chart for gender vs total amount
sale=data.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sale)
```

```python
#from the above chart we conclude that most of the buyer are female
```

```python
ax = sns.countplot(data = data, x = 'Age Group', hue = 'Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
age=data.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Age Group',y= 'Amount' ,data =age)
```

```python
#from the above chart we concluse that most of the buyer are from age group between 26-35 
```

```python
# total number of orders from top 10 states
total=data.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Orders',data=total)
```

```python
# total amount/sales from top 10 states
total=data.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(x='State',y='Amount',data=total)
```

```python
martial=sns.countplot(x='Marital_Status',data=data)
for bars in martial.containers:
    martial.bar_label(bars)
```

```python
datah=data.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Marital_Status',y='Amount',data=datah,hue='Gender')
```

```python
#we conclude that most of the buyer are married women
```

```python
occu=sns.countplot(x='Occupation',data=data)
for bars in occu.containers:
    occu.bar_label(bars)
```

```python
occ=data.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Occupation',y='Amount',data=occ)
```

```python
#from the above chart we conclude that most of the buyer are from it sector abd healathcare sector
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = data, x = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = data.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```

```python
#from the above we conclude that most og the buyer buy food and clothing 
```
